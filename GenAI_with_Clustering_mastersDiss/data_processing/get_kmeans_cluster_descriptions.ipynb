File name: get_kmeans_cluster_descriptions 

Description: After clustering was implemented on the data this file was created to group the desciptions of the gin based on their assigned clusters and then save a csv containing them both.

```python
import json
import pandas as pd
```

```python
df = pd.read_csv('../data/unlabelled_data.csv')

with open('../data/json/KMeans_labels.json', 'r') as f:
    kmeans_labels = json.load(f)
```

```python
#save labels on a csv
df["KMeans_labels"] = kmeans_labels["labels"]
df.to_csv("../data/unlabelled_data.csv", index=False)
```

```python
#save a grouped csv with descriptions and labels
descriptions_labels = df[["KMeans_labels", "Tasting Note (MoM)"]]
descriptions_labels.to_csv("../data/labelsAndDescirptions.csv", index=False)
```
