File name: data_preprocess

Description: This file aligns rows and headers to be in the correct position within the dataframe, then removes unwanted rows and does some light visualisation.

```python
import pandas as pd

df = pd.read_excel("../data/refinded_Gin(Nov2022).xlsm")
df.columns.__len__()
```

exploring positioning of rows and columns

```python
for index, x in enumerate(df.columns):
    print(index,"-",x)
```

align rows and headers

```python
df.columns = df.iloc[7]
```

```python
df.drop([0,1,2,3,4,5,6,7,560,561,562], inplace=True)
df.head()
```

```python
delete_rows = df.loc[df['LD (F)'] == 1].index.tolist()
delete_rows.extend(df.loc[df['Misc. (F)'] == 1].index.tolist())
print(delete_rows)
df.drop(delete_rows, inplace=True)
```

# removing unless columns

```python
for index, x in enumerate(df.isna().sum()):
    print(index,"-",df.columns[index],"\n---",x)
```

## visualise outliers

```python
plot = df["Total Known Botanicals"].plot.box(vert=False, rot=45)
plot.set_yticks([1], ["Botanicals"])
plot.set_xlabel("Amount")
plot.set_title("Spread of Botanicals Within Each Recipe")
```

```python
df[["Total Known Botanicals", "Spice (F)"]].describe()
```

```python
with open("../data/removed_coloumns.txt") as f:
    drop_these = f.read().splitlines()
df.drop(drop_these, axis=1, inplace=True)
```
